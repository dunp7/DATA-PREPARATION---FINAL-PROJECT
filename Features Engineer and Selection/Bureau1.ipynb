# Features Engineering and Selection

### Time related information

```python
time = ['DAYS_CREDIT','CREDIT_DAY_OVERDUE', 'DAYS_CREDIT_ENDDATE', 'DAYS_ENDDATE_FACT', 'DAYS_CREDIT_UPDATE']
df_time = df[time]
df_time
```

#### drop the DAYS_ENDDATE_FACT column
#### convert numbers to positive numbers

```python
df = df.drop('DAYS_ENDDATE_FACT', axis=1)
df['DAYS_CREDIT'] = df['DAYS_CREDIT'].fillna(0).astype(int).abs()
df['DAYS_CREDIT_ENDDATE'] = df['DAYS_CREDIT_ENDDATE'].fillna(0).astype(int).abs()
df['DAYS_CREDIT_UPDATE'] = df['DAYS_CREDIT_UPDATE'].fillna(0).astype(int).abs()
```

### Amount related information

```python
amount = ['AMT_CREDIT_MAX_OVERDUE', 'CNT_CREDIT_PROLONG', 'AMT_CREDIT_SUM', 'AMT_CREDIT_SUM_DEBT', 'AMT_CREDIT_SUM_LIMIT', 'AMT_CREDIT_SUM_OVERDUE', 'AMT_ANNUITY']
df_amount = df[amount]
df_amount
```

### Other information

```python
other = ['SK_ID_BUREAU', 'CREDIT_ACTIVE', 'CREDIT_CURRENCY', 'CREDIT_TYPE', 'SK_ID_CURR']
df_other = df[other]
df_other
```
