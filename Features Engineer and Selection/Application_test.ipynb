# Features Engineering and Selection

## The Effect of all the features to the target columns
* This data set is all about the information related to clients
* I can divide it into small parts
    + Personal Information (13)
    + Working and Education Information (4) (v)

    + Contact Information (7) (v)
    + Region Information (9) (v)

### Personal Information

```python
pinfo_col = ['NAME_CONTRACT_TYPE','CODE_GENDER', 'CNT_CHILDREN', 'NAME_HOUSING_TYPE',
             'NAME_TYPE_SUITE','NAME_FAMILY_STATUS','CNT_FAM_MEMBERS',  'AMT_INCOME_TOTAL',
            'DAYS_BIRTH', 'DAYS_EMPLOYED','WEEKDAY_APPR_PROCESS_START',
            'HOUR_APPR_PROCESS_START','DAYS_REGISTRATION','DAYS_ID_PUBLISH']
pinfo = df1[pinfo_col]
```

```python
# The infor relating to time
# Day related to the loan
ltime_pinfo = pinfo[['WEEKDAY_APPR_PROCESS_START','HOUR_APPR_PROCESS_START','DAYS_REGISTRATION','DAYS_ID_PUBLISH']]
```

```python
# Day related to the Clients
ctime_pinfo = pinfo[['DAYS_BIRTH', 'DAYS_EMPLOYED']]
ctime_pinfo             
```

##### Conclusion drop all the day relaing to the loan and convert DAYS_BIRTH and DAYS_EMPLOYED into AGE, and YEARS_EMPLOYED

```python
df1 = df1.drop(columns=ltime_pinfo.columns,axis = 0)
df1['DAYS_BIRTH'] = df1['DAYS_BIRTH'].apply(lambda x: int((x * -1) / 365))
df1['DAYS_EMPLOYED'] = df1['DAYS_EMPLOYED'].apply(lambda x: int((x * -1) / 365))
df1 = df1.rename({'DAYS_BIRTH': 'AGE', 'DAYS_EMPLOYED' : 'YEARS_EMPLOYED'}, axis=1)
```

### Working and Education Information

##### Conclude : Keep Working adn Education Information

### Asset information

```python
ainfo_col = ['AMT_ANNUITY','AMT_CREDIT', 'AMT_GOODS_PRICE','FLAG_OWN_CAR','FLAG_OWN_REALTY', 'OWN_CAR_AGE']
ainfo = df1[ainfo_col]
ainfo
```

#### Conclusion : Keep only AMT_ANNUITY in Asset Information

```python
df1 = df1.drop(columns=ainfo_col[1:],axis = 0)
```

### Contact Information

```python
cinfo_col = ['FLAG_MOBIL', 'FLAG_EMP_PHONE', 'FLAG_WORK_PHONE', 'FLAG_CONT_MOBILE', 'FLAG_PHONE', 'FLAG_EMAIL','DAYS_LAST_PHONE_CHANGE']
cinfo = df1[cinfo_col]
cinfo
```

#### Conclude -> Remove all the contact features

### Region Information (v)

```python
rinfo_col = ['REGION_RATING_CLIENT','REGION_RATING_CLIENT_W_CITY','REG_REGION_NOT_LIVE_REGION',
            'REG_REGION_NOT_WORK_REGION','LIVE_REGION_NOT_WORK_REGION','REG_CITY_NOT_LIVE_CITY',
            'REG_CITY_NOT_WORK_CITY','LIVE_CITY_NOT_WORK_CITY']
rinfo = df1[rinfo_col]
rinfo
```

##### Conclude: This information is not useful at all -> remove region information

### Housing Information

### Document Information

```python
start = df1.columns.get_loc('FLAG_DOCUMENT_2')
end = df1.columns.get_loc('FLAG_DOCUMENT_21')
dinfo_col = list(df1.columns[start:end+1])
dinfo = df1[dinfo_col]
dinfo
```

### Bureau Information

```python
binfo_col = [
'AMT_REQ_CREDIT_BUREAU_HOUR',
'AMT_REQ_CREDIT_BUREAU_DAY',
'AMT_REQ_CREDIT_BUREAU_WEEK',
'AMT_REQ_CREDIT_BUREAU_MON',
'AMT_REQ_CREDIT_BUREAU_QRT',
'AMT_REQ_CREDIT_BUREAU_YEAR'
]
binfo = df1[binfo_col]
binfo
```


-> Conclude, tạo thêm 1 cột sum tất cả các giá trị vào. 

```python
## Add one column convert the value into day and sum all
df1['AMT_REQ_CREDIT_BUREAU_TOTAL_DAY'] = (
    df1['AMT_REQ_CREDIT_BUREAU_HOUR'] / 24 +
    df1['AMT_REQ_CREDIT_BUREAU_DAY'] +
    df1['AMT_REQ_CREDIT_BUREAU_MON'] * 30 +
    df1['AMT_REQ_CREDIT_BUREAU_YEAR'] * 365 +
    df1['AMT_REQ_CREDIT_BUREAU_WEEK'] * 7 +
    df1['AMT_REQ_CREDIT_BUREAU_QRT'] * 3 * 30
)
df1 = df1.drop(columns=binfo_col, axis = 1)
```

```python
df1 = df1.set_index('SK_ID_CURR')
```
