# Features Engineering and Selection

### Numerical data

```python
df = df.drop(columns=["DAYS_TERMINATION", "DAYS_LAST_DUE_1ST_VERSION",
                       "AMT_APPLICATION" ,"AMT_GOODS_PRICE",'RATE_DOWN_PAYMENT'], axis = 1)
```

```python
col = ['DAYS_DECISION','DAYS_FIRST_DRAWING','DAYS_FIRST_DUE','DAYS_LAST_DUE',\
    'CNT_PAYMENT','HOUR_APPR_PROCESS_START','SELLERPLACE_AREA']
df = df.drop(columns = col,axis = 1)
```

### Categorical column

```python
df = df.drop(columns='FLAG_LAST_APPL_PER_CONTRACT', axis = 1)
```

```python
col1 = ['NAME_CLIENT_TYPE']
col2= ['PRODUCT_COMBINATION']
col3 = ['NFLAG_INSURED_ON_APPROVAL','NFLAG_LAST_APPL_IN_DAY','NAME_TYPE_SUITE',
                     'WEEKDAY_APPR_PROCESS_START','NAME_CASH_LOAN_PURPOSE','CODE_REJECT_REASON','NAME_GOODS_CATEGORY']

df = df.drop(columns = col1,axis = 1)
df = df.drop(columns = col2,axis = 1)
df = df.drop(columns = col3,axis = 1)
```

```python
df
```

```python
num_data = df._get_numeric_data()
num_col = num_data.columns
cat_data = df[[col for col in df.columns if col not in num_data.columns]]
```

```python
# XNA is relevant to Null value. so we can remove it out of the features
for col in cat_data.columns:
    col_name  = f'{col.replace("NAME_", "")}'
    for value in df[col].unique():
        if value != 'XNA':
            new_col_name = f'{value}_{col_name}'
            df[new_col_name] = df[col].apply(lambda x: 1 if x == value else 0)

    df.drop(columns=col,inplace=True)

```

```python
df
```

#### Note

+ Because SK_ID_CURR in this table is duplicated, so we need to group by all of the data so that we can join with the application train to prevent wrong merge table


```python
num_data = num_data.groupby('SK_ID_CURR').agg({
    'AMT_ANNUITY': ['mean','min','max','sum'],
    'AMT_CREDIT': ['mean', 'sum', 'min', 'max'],
    'AMT_DOWN_PAYMENT' : ['mean', 'sum', 'min', 'max']
})
num_data.columns = ['_'.join(col).strip().upper() for col in num_data.columns.values]
num_data
```

```python
num_col = list(num_col)
num_col.remove('SK_ID_CURR')

cat_df = df[[col for col in df.columns if col not in num_col]]
cat_df = cat_df.groupby(['SK_ID_CURR']).agg(sum)
```

```python
final_prev_app = pd.merge(num_data,cat_df,left_index=True, right_index=True)
final_prev_app
```
