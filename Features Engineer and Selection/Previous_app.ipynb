# Features Engineering and Selection

### Numerical data

```python
num_data_col = df._get_numeric_data().columns
num_data_col = [col for col in num_data_col if col not in ['SK_ID_CURR', 'TARGET']]
num_data_col
```

```python
# df
# ratios and difference
df['APPLICATION_CREDIT_DIFF'] = df['AMT_APPLICATION'] - df['AMT_CREDIT']
df['APPLICATION_CREDIT_RATIO'] = df['AMT_APPLICATION'] / df['AMT_CREDIT']
df['CREDIT_TO_ANNUITY_RATIO'] = df['AMT_CREDIT']/df['AMT_ANNUITY']
df['DOWN_PAYMENT_TO_CREDIT'] = df['AMT_DOWN_PAYMENT'] / df['AMT_CREDIT']
df['AMT_INTEREST'] = df['CNT_PAYMENT'] * df['AMT_ANNUITY'] - df['AMT_CREDIT'] 
df['AMT_DECLINED'] = df['AMT_APPLICATION'] - df['AMT_CREDIT']

# Interest ratio on dfious application (simplified)
total_payment = df['AMT_ANNUITY'] * df['CNT_PAYMENT']
df['SIMPLE_INTERESTS'] = (total_payment/df['AMT_CREDIT'] - 1)/df['CNT_PAYMENT']

# make sure that the ratio is not infinity
df.replace([np.inf, -np.inf], np.nan,inplace=True)

# Days last due difference (scheduled x done)
df['DAYS_LAST_DUE_DIFF'] = df['DAYS_LAST_DUE_1ST_VERSION'] - df['DAYS_LAST_DUE']
```

```python
# sns.boxplot(df['SELLERPLACE_AREA'])
# From this boxplot we can see in this features, there exist one extreme outliers is 4.10^6, We fill this value by np.nan
#because we don't want to remove any rows
df['SELLERPLACE_AREA'] = df['SELLERPLACE_AREA'].replace(4000000,np.nan)
```

```python

#label encoding the categorical variables

# ordinal columns
ord_col = ['NAME_CONTRACT_STATUS','NAME_YIELD_GROUP','FLAG_LAST_APPL_PER_CONTRACT']
encode_contract_status = {'Approved': 0, 'Refused' : 3, 'Canceled' : 2, 'Unused offer' : 1,'XNA': 4}
df['NAME_CONTRACT_STATUS'] = df['NAME_CONTRACT_STATUS'].map(encode_contract_status)
encode_yield_group = {'XNA': 0, 'low_action': 1, 'low_normal': 2,'middle': 3, 'high': 4}
df['NAME_YIELD_GROUP'] = df['NAME_YIELD_GROUP'].map(encode_yield_group)
encode_appl_per_contract_last = {'Y':1, 'N':0}
df['FLAG_LAST_APPL_PER_CONTRACT'] = df['FLAG_LAST_APPL_PER_CONTRACT'].map(encode_appl_per_contract_last)

# norminal columns

for col in [col for col in cat_col if col not in ord_col]:
        encoding_dict = dict([(j,i) for i,j in enumerate(df[col].unique(),1)])
        df[col] = df[col].map(encoding_dict) 

```

#### Note

+ Because SK_ID_CURR in this table is duplicated, so we need to group by all of the data so that we can join with the application train to prevent wrong merge table


```python
num_data_col
```

```python
num_data = df[num_data_col].groupby('SK_ID_CURR').agg({
    # amount
    'AMT_ANNUITY': ['mean','min','max','sum'],
    'AMT_APPLICATION' : ['mean','min','max','sum'],
    'AMT_CREDIT': ['mean', 'sum', 'min', 'max'],
    'AMT_DOWN_PAYMENT' : ['mean', 'sum', 'min', 'max'],
    'AMT_GOODS_PRICE': ['mean', 'sum', 'min', 'max'],
    'AMT_INTEREST': ['mean', 'sum', 'min', 'max'],
    'APPLICATION_CREDIT_DIFF':['mean', 'sum', 'min', 'max'],
    'AMT_DECLINED':['mean', 'sum', 'min', 'max'],
    
    # ratio
    'APPLICATION_CREDIT_RATIO':['mean','min', 'max'],
    'CREDIT_TO_ANNUITY_RATIO' : ['mean','min', 'max'],
    'DOWN_PAYMENT_TO_CREDIT':['mean','min', 'max'],
    'SIMPLE_INTERESTS':['mean','min', 'max'],
    'RATE_DOWN_PAYMENT' : ['mean','max'],

    # day
    'DAYS_LAST_DUE_DIFF': ['mean', 'sum', 'min', 'max'],
    'DAYS_FIRST_DRAWING' : ['mean','max'],
    'DAYS_FIRST_DUE' : ['mean','max'],
    'DAYS_LAST_DUE_1ST_VERSION' : ['mean'],
    'DAYS_LAST_DUE' : ['mean'],
    'DAYS_TERMINATION' : ['mean','max'],

    # flag
    'NFLAG_INSURED_ON_APPROVAL': ['mean','last'],
    'NFLAG_LAST_APPL_IN_DAY' : ['mean','sum'],
    
    # count
    'CNT_PAYMENT': ['sum','mean','max'],
    'SELLERPLACE_AREA' : ['mean','max','min']
})
num_data.columns = ['_'.join(col).strip().upper() for col in num_data.columns.values]
num_data
```

```python
cat_col.append('SK_ID_CURR')
cat_col
```

```python
cat_data = df[cat_col].groupby(['SK_ID_CURR']).agg({
    'NAME_CONTRACT_TYPE' : ['mean','last'],
    'WEEKDAY_APPR_PROCESS_START': ['mean','max','min'],
    'FLAG_LAST_APPL_PER_CONTRACT': ['mean','sum'],
    'NAME_CASH_LOAN_PURPOSE': ['mean','last'],
    'NAME_CONTRACT_STATUS': ['mean','max','last'],
    'NAME_PAYMENT_TYPE': ['mean', 'last'],
    'CODE_REJECT_REASON': ['mean', 'last'],
    'NAME_TYPE_SUITE' : ['mean','last'],
    'NAME_CLIENT_TYPE' : ['mean','last'],
    'NAME_GOODS_CATEGORY' : ['mean','last'],
    'NAME_PORTFOLIO' : ['mean','last'],
    'NAME_PRODUCT_TYPE' : ['mean','last'],
    'CHANNEL_TYPE' : ['mean','last'],
    'NAME_SELLER_INDUSTRY': ['mean','last'],
    'NAME_YIELD_GROUP': ['mean','last'],
    'PRODUCT_COMBINATION': ['mean', 'last']
})

cat_data.columns = ['_'.join(col).strip().upper() for col in cat_data.columns.values]
cat_data
```

```python
final_prev_app = pd.merge(num_data,cat_data,left_index=True, right_index=True)
final_prev_app
```
